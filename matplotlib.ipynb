## Art of Plotting with Matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
# Creates a new figure, if we pass a number figure(3) and if a figure with that number exists, it gets returned

fig = plt.figure(figsize=(3, 3), facecolor="skyblue") # a new figure with no axes

fig.suptitle("My first figure")

# axes is a subplot, and we create a subplot in the above figure
ax = fig.add_subplot()
ax.set_title("subplot A", loc="left")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")

# lets create another subplot --> doing this overwrites exising ax
# ax2 = fig.add_subplot()
# ax2.set_title("subplot B")
```

### OOP Way of Plotting
Here we explictly create a figure and use the subplot created by figure for plotting

```python
import numpy as np

x = np.linspace(0, 2, 100)


# create a subplot 
fig, ax = plt.subplots()
""" When we don't pass grid size, a single subplot is returned i.e. ax """

# We will use this ax to create plot
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")

""" Since we have created multiple line plots in the same axes, we need
a way to distinguish among them, so we will use legend for the labels we have added 
to each line plot
"""

ax.legend()

ax.set_xlabel("x axis")
ax.set_ylabel("y axis")

ax.set_title("Comparative Line Graph")

plt.show()
#REVIEW: 
""" What happens when we call plt.show()?? """

```

### 2. Non-OOP Style of Plotting
using plt and implicitly creating figure and plotting

```python

plt.figure()
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")

plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```
